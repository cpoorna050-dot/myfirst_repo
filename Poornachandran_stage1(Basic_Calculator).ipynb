```python
def basic_calculator():
    Num1 = int(input("entre number 1:"))
    Num2 = int(input("entre number 2:"))
    operators = input("entre any one operator(+,-,*,/):")
    if operators == "+":
        Result = sum(Num1,Num2)
        print(Result)
    elif operators == "-":
        Result = Num1 - Num2
        print (Result)
    elif operators == "*":
        Result = Num1 * Num2
        print (Result)
    elif operators == "/":
        if Num2 == 0:
            print("Cannot divisible by 0")
        else:
            Result = Num1 / Num2
            print (Result)
basic_calculator()
```
