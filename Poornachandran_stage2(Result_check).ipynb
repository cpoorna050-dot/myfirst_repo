```python
def Result_check(): #stage 2
    input1 = int(input("1st input:"))
    input2 = int(input("2nd input:"))
    Result = input1 - input2
    if Result > 0 :
        print("Positive")
    elif Result < 0 :
        print("Negative")
    else:
        print("Zero")
Result_check()
```
