```python
def Student_Grade_calculator(): #stage3
    import math
    Student_name = input("Name of the student:")
    Sub1 = int(input("Entre subject 1 mark:"))
    Sub2 = int(input("Entre subject 2 mark:"))
    Sub3 = int(input("Entre subject 3 mark:"))
    total = Sub1+Sub2+Sub3
    percentage = (total/300)*100
    print("Student name:", Student_name)
    print("Total marks:",total)
    print("percentage:",percentage)
    if percentage >= 75:
        print("Grade:A")
    elif percentage >= 60:
        print("Grade:B")
    elif percentage >= 40:
        print("Grade:C")
    else:
        print("Grade:F")
Student_Grade_calculator()
```
